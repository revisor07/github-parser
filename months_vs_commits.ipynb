```python
import datetime
import numpy
import os
from matplotlib import pyplot as plt
import matplotlib
import pandas
```

```python
#file path
path = os.path.join(os.path.expanduser('~'), 'Documents', 'TalentGames', 
  'data','facebook','test.txt')
commits = []
table = {}
date_list = []
commit_list = []
with open(path) as s:
  commits = s.readlines()
```

```python
for i in commits:
    line = i.split("\t")
    time_stamp = pandas.to_datetime(line[5])
    year = str(time_stamp.year)
    month = str(time_stamp.month)
    day = str(time_stamp.day)
    if(len(month) == 1):
        month = "0" + month
    if(len(day) == 1):
        day = "0" + day
    date = year + "-" + month + "-" + day
    if(time_stamp.year == 2017 or time_stamp.year == 2018):
        if(date not in table.keys()):
            table[date] = 0
        table[date] += 1
for key, value in sorted(table.items()):
    date_list.append(key)
    if(value > 5000):
        value = 400
    commit_list.append(value)
```

```python
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 30
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.plot(date_list, commit_list, c="red", alpha=0.5)
plt.xlabel("Day")
plt.ylabel("Commits")
plt.legend(loc=2)
plt.show()
```
